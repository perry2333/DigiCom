```python
# 定义要读取的文件名和要保存的文件名
input_filename = 'rx151024_real.m'  # 输入文本文件名
output_python_data = 'formatted_python_data.txt'  # 输出处理后的数据文本文件

# 读取输入文件
with open(input_filename, 'r', encoding='utf-8') as file:
    data = file.read()

# 去掉换行符，替换 'j' 为 'i'
data = data.replace('\n', ' ').replace('j', 'i')s

# 将数据从 "[...]" 内部提取出来，并按逗号分隔
data = data.split('=')[1].strip()  # 获取等号后的数据
data = data[1:-1]  # 去掉外层的方括号

# 使用逗号分隔每个元素
data = data.replace(') ', '),')  # 在复数之间插入逗号

#data = data.replace('i*(', '').replace(')','*i')  

# 保存处理好的数据到文件，之后可以分批读取
with open(output_python_data, 'w', encoding='utf-8') as file:
    file.write(f'[{data}]')

print(f"数据已保存为可读取的文本文件：{output_python_data}")
```

```python
import numpy as np
from scipy.io import savemat

# 文件名定义
input_python_data = 'formatted_python_data.txt'  # 输入已格式化的Python文本数据
output_real_mat_template = 'output_real_data_part{}.mat'
output_imag_mat_template = 'output_imag_data_part{}.mat'
output_mat_template = 'output_real/imag_data_part{}.mat'  # 输出 .mat 文件模板

# 定义每一批处理的数据量
total_data_size = 1228800  # 总数据量
num_batches = 1  # 分10批处理
batch_size = total_data_size // num_batches  # 每批的数据量

# 打开并逐行读取数据
with open(input_python_data, 'r', encoding='utf-8') as file:
    # 我们假设数据在方括号中： [...], 因此我们去掉开头和结尾的方括号
    data = file.read().strip()[1:-1]  # 去掉 '[' 和 ']'

# 将数据拆分为复数的字符串列表，每个复数以逗号分隔
data.replace(" + ","").replace("(","").replace(")","").replace(' ','')
data_split = data.split('),')

# 因为每个复数是形如 'x + i*(y)'，处理一下每个字符串，修正格式并添加逗号
data_split = [item + ')' if not item.endswith(')') else item for item in data_split]

data_split_real = []
data_split_imag = []
for l in data_split:
    l = l.replace(" + ","").replace("(","").replace(")","")
    Ls = l.split("i*")
    if len(Ls) == 2:
        real_part = Ls[0].strip()  # 第一个是实部
        imag_part = Ls[1].strip()  # 第二个是虚部

        # 将实部和虚部分别存入对应的列表
        data_split_real.append(real_part)
        data_split_imag.append(imag_part)

data_array_real = np.array(data_split_real)
data_array_imag = np.array(data_split_imag)
'''
savemat(output_real_mat_template.format(1), {'rx151024_real': data_array_real})
savemat(output_imag_mat_template.format(1), {'rx151024_imag': data_array_imag})
'''

# 分批处理并保存为 .mat 文件
for i in range(num_batches):
    start_idx = i * batch_size
    end_idx = (i + 1) * batch_size
    data_batch_real = data_split_real[start_idx:end_idx]
    data_batch_imag = data_split_imag[start_idx:end_idx]  # 提取一批数据

    # 使用 eval() 将字符串批量转化为复数
    data_batch_real_eval = [eval(item) for item in data_batch_real]
    data_batch_imag_eval = [eval(item) for item in data_batch_imag]  # 每个字符串转为 Python 复数

    # 将列表转为 numpy 数组
    data_array_real = np.array(data_batch_real_eval)
    data_array_imag = np.array(data_batch_imag_eval)

    # 保存每一批为 .mat 文件，文件名为 output_data_part1.mat, output_data_part2.mat, etc.
    savemat(output_real_mat_template.format(i+1), {'rx151024_real': data_array_real})
    savemat(output_imag_mat_template.format(i+1), {'rx151024_imag': data_array_imag})

    print(f"第 {i+1} 批数据已保存为 {output_mat_template.format(i+1)}")

print("所有批次数据保存完成。")
```
